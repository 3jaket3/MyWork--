```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Wine = pd.read_csv('Wine.csv')
Wine = Wine.iloc[:,0:12]
from pandas.api.types import is_numeric_dtype

for col in Wine.columns:
    if is_numeric_dtype(Wine[col]):
        print('%s: ' % (col))
        print('\t Mean = %.2f' % Wine[col].mean())
        print('\t std = %2f' % Wine[col].std())
        print('\t min = %2f' % Wine[col].min())
        print('\t max = %2f' % Wine[col].max())
        
Wine['Class'].value_counts()
Wine.describe(include='all')
print('Cov: ')
Wine.cov()
print('correlation: ')
Wine.corr()      



```

```python
Wine2 = Wine.loc[:,['Flavanoids',"Total phenols"]]
Z = (Wine2 - Wine2.mean())/Wine2.std();

print('Number of rows before discarding outliers = %d' % (Z.shape[0]))
Z2 = Z.loc[((Z >-1).sum(axis=1)==2) & ((Z<=1).sum(axis=1)==2),:]
print('Number of rows after discarding outliers = %d' % (Z2.shape[0]))
Z2 = Z2**(Wine2.std()) + Wine2.mean()    
```

```python
Flav2 = Z2.Flavanoids
Phen2 = Z2["Total phenols"]
plt.scatter(Flav2,Phen2)
```
